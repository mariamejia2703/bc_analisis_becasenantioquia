```python
def detectar_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
    return outliers
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#se define df entre las variables numericas
data = df[['edad','convocatoria']]
data
```

```python
#Se crea df resultante de la data sin outliers de edad
Q1 = data['edad'].quantile(0.25)
Q3 = data['edad'].quantile(0.75)

IQR = Q3-Q1

piso=Q1-1.5*IQR
TECHO=Q3+1.5*IQR

data_sin_outliers = data[(data['edad']>=piso) & (data['edad']<=TECHO)]
data_sin_outliers
```

```python
#estándariza los datos
scalador = StandardScaler()
data_escalada = scalador.fit_transform(data_sin_outliers)
data_escalada
```

```python
#WCSS calcula la distancia entre grupos
WCSS = []
for i in range(1,15):
    km = KMeans(n_clusters=i)
    km.fit(data_escalada)
    WCSS.append(km.inertia_)
```

```python
WCSS
```

```python
#aplicamos la figura (método del codo)
plt.figure(figsize=(10,5))
plt.plot(range(1,15),WCSS)
plt.show()
```

```python
km = KMeans(n_clusters=3)
km.fit(data_escalada)
```

```python
km.fit_predict(data_escalada)
```

```python
from sklearn.metrics import silhouette_score
```

```python
#Se ejecuta el método de silueta,El análisis de silueta se puede utilizar para examinar la cantidad de separación entre los grupos.
silueta=[]
for i in range(2,15):
    km = KMeans(n_clusters=i)
    clusters = km.fit_predict(data_escalada)
    silueta_prom=silhouette_score(data_escalada,clusters)
    silueta.append(silueta_prom)

silueta
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(2,15),silueta, marker='o')
plt.title("Coeficiente de silueta para diverson valores de K")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Coeficiente de silueta")
plt.show()
```

```python
k_optimo = 3
modelo_km = KMeans(n_clusters=k_optimo)
clusters = modelo_km.fit_predict(data_escalada)
```

```python
#se crea la columna cluster y se asigna la variable cluster
data_sin_outliers['cluster']=clusters
data_sin_outliers
```

```python
data_sin_outliers['cluster'].unique()
```

```python
plt.figure(figsize=(12,8))
for cluster in sorted(data_sin_outliers['cluster'].unique()):
    plt.scatter(
        data_sin_outliers.loc[data_sin_outliers['cluster']==cluster,'edad'],
        data_sin_outliers.loc[data_sin_outliers['cluster']==cluster,'convocatoria'],
        label=f'Cluster {cluster}'
    )
```

```python
scalador.inverse_transform(modelo_km.cluster_centers_)
```

```python
plt.figure(figsize=(12, 8))
for cluster in sorted(data_sin_outliers["cluster"].unique()):
  plt.scatter(
      data_sin_outliers.loc[data_sin_outliers["cluster"] == cluster,"edad"],
      data_sin_outliers.loc[data_sin_outliers["cluster"] == cluster,"convocatoria"],
      label=f"Cluster{cluster}"
  )

centroides = scalador.inverse_transform(modelo_km.cluster_centers_)
plt.scatter(centroides[:, 0], centroides [:,1], s=100, c='red', marker='x',label='Centroides')
plt.show()
```

```python
# Filtramos los diferentes grupos étnicos
grupos_etnicos = distribucion['grupo_etnico'].unique()

# Crear un gráfico por cada grupo étnico
for grupo in grupos_etnicos:
    # Filtrar datos por el grupo étnico actual
    datos_grupo = distribucion[distribucion['grupo_etnico'] == grupo]

    # Crear una figura para cada gráfico
    plt.figure(figsize=(12, 8))

    # Gráfico de barras agrupadas para las tres variables (género, estrato, victima_conflicto_armado)
    sns.barplot(x='estrato', y='conteo', hue='género', data=datos_grupo, palette='Set2')

    # Título del gráfico
    plt.title(f'Distribución por Género, Estrato y Víctima del Conflicto Armado\nGrupo Étnico: {grupo}', fontsize=16)
    plt.xlabel('Estrato', fontsize=12)
    plt.ylabel('Número de Beneficiarios', fontsize=12)

    # Mostrar leyenda
    plt.legend(title='Género')

    # Ajustar la visualización
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()

    # Gráfico de barras agrupadas para víctimas del conflicto armado
    plt.figure(figsize=(12, 8))
    sns.barplot(x='estrato', y='conteo', hue='victima_del_conflicto_armado', data=datos_grupo, palette='coolwarm')

    # Título del gráfico
    plt.title(f'Distribución de Víctimas del Conflicto Armado por Estrato\nGrupo Étnico: {grupo}', fontsize=16)
    plt.xlabel('Estrato', fontsize=12)
    plt.ylabel('Número de Beneficiarios', fontsize=12)

    # Mostrar leyenda
    plt.legend(title='Víctima del Conflicto Armado')

    # Ajustar la visualización
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()
```


**12. ¿Qué grupo etario predomina entre los beneficiarios x genero?**

```python
# Agrupar por 'género' y calcular el rango de edades por grupo
rango_edades_por_genero = df.groupby('género')['edad'].agg(['min', 'max'])

# Calcular el rango (diferencia entre max y min) para cada grupo
rango_edades_por_genero['rango'] = rango_edades_por_genero['max'] - rango_edades_por_genero['min']

# Mostrar el resultado
print(rango_edades_por_genero)
```

```python
# Crear un gráfico de líneas para el rango de edades por género
plt.figure(figsize=(8, 6))

# Graficar el rango de edades con un gráfico de líneas
sns.lineplot(data=rango_edades_por_genero, x=rango_edades_por_genero.index, y='rango', marker='o', color='purple')

# Añadir título y etiquetas
plt.title('Rango de Edad por Género', fontsize=16)
plt.xlabel('Género', fontsize=12)
plt.ylabel('Rango de Edad', fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
